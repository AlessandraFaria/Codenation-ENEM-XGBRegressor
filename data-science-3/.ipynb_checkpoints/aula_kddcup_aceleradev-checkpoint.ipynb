### Verificando a consistência da base

```python
#Criando um dataframe auxliar para analisar a consistencia das variaveis
cons = pd.DataFrame({'colunas' : df.columns,
                    'tipo': df.dtypes,
                    'missing' : df.isna().sum(),
                    'size' : df.shape[0],
                  'unicos': df.nunique()})
cons['percentual'] = round(cons['missing'] / cons['size'],2)
```

```python
cons.head()
```

```python
cons.percentual.plot.hist( bins = 5)
```

#### Removendo as colunas com muitos dados faltantes

```python
print('Contagem de colunas com ATÉ 20% de dados faltantes', cons[cons.percentual < 0.2].shape[0])
print('Contagem de colunas com 0% de dados faltantes',  cons[cons.percentual == 0].shape[0])
```

```python
cons[cons.percentual < 0.2]['tipo'].value_counts()
```

```python
cons[cons.percentual == 0]['tipo'].value_counts()
```

```python
cons['completa'] = ['completa' if x == 0 else 'faltante' for x in cons['percentual']]
cons['vinte'] = ['vinte' if x < 0.20 else 'remover' for x in cons['percentual']]
```

```python
mantem = list(cons[cons['vinte'] == 'vinte']['colunas'])
df = df[mantem]
```

```python
df.columns
```

```python
cons.columns
```

```python
colunas_numericas = list(cons[((cons['tipo'] != 'object') &
                              (cons['vinte'] == 'vinte'))]['colunas'])
colunas_categoricas = list(cons[((cons['tipo'] == 'object') &
                                (cons['vinte'] == 'vinte'))]['colunas'])
colunas_poucascategorias = list(cons[((cons['tipo'] == 'object')  &
                                     (cons['vinte'] == 'vinte') & 
                                    (cons['unicos'] < 6))]['colunas'])
```

#### Tratando os dados faltantes

```python
medias = {}
for coluna in colunas_numericas:
    aux1= {'media_'+coluna : df[coluna].mean()}
    medias.update(aux1)
```

```python
desvios = {}
for coluna in colunas_numericas:
    aux1= {'desvio_'+coluna : df[coluna].std()}
    desvios.update(aux1)
```

```python
for coluna in colunas_numericas:
    df[coluna] = df[coluna].fillna(medias['media_'+coluna])
```

```python
df[colunas_numericas].isna().sum()
```

```python
print(cons[(cons['tipo'] == 'object') &
                           (cons['vinte'] == 'vinte')&
                           (cons['completa'] == 'faltante')])
possivel_tratamento = list(cons[(cons['tipo'] == 'object') &
                           (cons['vinte'] == 'vinte')&
                           (cons['completa'] == 'faltante')]['colunas'])
```

```python
df.drop(columns=possivel_tratamento, inplace=True)
```

```python
print(cons[(cons['tipo'] == 'object') &
                        (cons['vinte'] == 'vinte')& 
                        (cons['missing'] == 0) &
                        (cons['unicos'] < 10)])
categoricas_modelo = list(cons[(cons['tipo'] == 'object') &
                        (cons['vinte'] == 'vinte')& 
                        (cons['missing'] == 0)&
                              (cons['unicos'] < 10)]['colunas'])
```

```python
numericas_modelo = colunas_numericas.copy()
```

#### Tratando as variavéis para modelagem

```python
#Aplicando o zscore nas variaveis númericas
for coluna in numericas_modelo:
    df[coluna] = (df[coluna] - medias['media_'+coluna] ) / desvios['desvio_'+coluna]
```

```python
df[numericas_modelo].describe().round(2)
```

```python
from sklearn.model_selection import train_test_split
X = pd.get_dummies(df[numericas_modelo+categoricas_modelo])
y = df['labels']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1 )
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=1)
print("Quantidade Observações:\nTreino: {tr}\nTeste: {te}\nValidação: {v}".format(tr = X_train.shape[0],
                                                                                 te = X_test.shape[0],
                                                                                 v = X_val.shape[0]))
```

```python
X_train.shape
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
```

```python
X_resampled.shape
```

```python
from tpot import TPOTClassifier
```

```python
tp = TPOTClassifier(generations=5, population_size=20, cv=5,
                                    random_state=42, verbosity=2)
```

```python
tp.fit(X_resampled, y_resampled)
```

```python
#exportanod o modelo
tp.export('modelo.py')
with open('modelo.py') as pipe:
    print("".join(pipe.readlines()))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(bootstrap=True, criterion="gini", max_features=0.7000000000000001, min_samples_leaf=3, min_samples_split=5, n_estimators=100)
```

```python
rf.fit(X_resampled, y_resampled)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred=y_pred, y_true=y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_pred=y_pred, y_true=y_test)
```

```python
sns.heatmap(confusion_matrix(y_pred=y_pred, y_true=y_test), annot = True)
```
